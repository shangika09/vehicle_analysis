```python
import matplotlib.pyplot as plt

data = df.groupby("year")["ev_sales"].sum()
plt.plot(data.index, data.values)
plt.title("EV Sales by Year")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = df.groupby("year")["petrol_car_sales"].sum()
plt.bar(data.index, data.values)
plt.title("Petrol Sales by Year")
plt.show()
```

```python
data = df.groupby("country")["ev_sales"].sum().head(5)
plt.pie(data.values, labels=data.index, autopct="%1.1f%%")
plt.title("Top 5 Countries EV Sales")
plt.show()
```
